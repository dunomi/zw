```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#https://matplotlib.org/tutorials/introductory/pyplot.html
#https://jakevdp.github.io/PythonDataScienceHandbook/04.01-simple-line-plots.html
```

```
def annotate_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)


fig = plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))

#Wykres 1
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
ax1.axis([0, 6, 0, 20])

#Wykres 2
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

ax2.scatter('a', 'b', c='c', s='d', data=data)

#Wykres 3
ax3.bar(['group_a', 'group_b', 'group_c'], [1, 10, 100])

#Wykres 4
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
ax4.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

#Wykres 5
x = np.linspace(0, 10, 1000)
ax5.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
ax5.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
ax5.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
ax5.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
ax5.plot(x, np.sin(x - 5), color='chartreuse');

annotate_axes(fig)

plt.show()
```

```
import random
import matplotlib.pyplot as plt
from matplotlib import style

def create_plots():
    xs = []
    ys = []

    for i in range(10):
        x = i
        y = random.randrange(10)

        xs.append(x)
        ys.append(y)
    return xs, ys

create_plots()
```
