```
# Z powodu użycia innej biblioteki w zadaniu 9 i jej niespójności z poleceniem subplot2grid 
# użyliśmy przykładowych wykresów ze stron podanych poniżej

# Źródło wykresów 
#https://matplotlib.org/tutorials/introductory/pyplot.html
#https://jakevdp.github.io/PythonDataScienceHandbook/04.01-simple-line-plots.html
#https://jakevdp.github.io/PythonDataScienceHandbook/04.02-simple-scatter-plots.html

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```
def annotate_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)


fig = plt.figure()
ax1 = plt.subplot2grid((3, 8), (0, 0), colspan=2)
ax2 = plt.subplot2grid((3, 8), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 8), (1, 2), colspan=3, rowspan=2)
ax4 = plt.subplot2grid((3, 8), (2, 0), colspan=2)
ax5 = plt.subplot2grid((3, 8), (0, 2), colspan=3)
ax6 = plt.subplot2grid((3, 8), (0, 5), colspan=3, rowspan=3)


ax1.set_title('plot 1')
ax2.set_title('plot 2')
ax3.set_title('plot 3')
ax4.set_title('plot 4')
ax5.set_title('plot 5')
ax6.set_title('plot 6')

ax1.set_ylabel('y-label')
ax2.set_ylabel('y-label')
ax4.set_ylabel('y-label')
ax5.set_ylabel('y-label')
ax6.set_ylabel('y-label')

ax1.set_xlabel('plot 1 x-label')
ax2.set_xlabel('plot 2 x-label')
ax3.set_xlabel('plot 3 x-label')
ax4.set_xlabel('plot 4 x-label')
ax5.set_xlabel('plot 5 x-label')
ax6.set_xlabel('plot 6 x-label')

#Wykres 1
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
ax1.axis([0, 6, 0, 20])

#Wykres 2
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

ax2.scatter('a', 'b', c='c', s='d', data=data)

#Wykres 3
ax3.bar(['group_a', 'group_b', 'group_c'], [1, 10, 100])

#Wykres 4
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
ax4.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

#Wykres 5
x = np.linspace(0, 10, 1000)
ax5.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
ax5.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
ax5.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
ax5.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
ax5.plot(x, np.sin(x - 5), color='chartreuse');

#Wykres 6
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    ax6.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))




#annotate_axes(fig)
plt.tight_layout()
plt.show()
```
